# Neural Architecture Primer

## Objective
Understand the fundamental types of neural network architectures and how to choose the right model based on data type and problem statement.

## Why Deep Learning Now?

Deep Learning became practical due to the convergence of three key factors:

- **Large-scale labeled data**: Availability of massive datasets from digital platforms.
- **Hardware advancements**: GPUs and high-performance processors enabling faster computation.
- **Improved algorithms & frameworks**: TensorFlow, Keras, and optimization techniques simplified implementation.

These factors transformed neural networks from theoretical models into real-world solutions.

## Types of Neural Network Architectures

### 1. Artificial Neural Networks (ANN / MLP)
- Used for structured/tabular data
- Captures non-linear relationships
- Applications: regression, classification

### 2. Convolutional Neural Networks (CNN)
- Designed for image/grid-like data
- Detects spatial patterns (edges → objects)
- Applications: image classification, object detection

### 3. Recurrent Neural Networks (RNN / LSTM / GRU)
- Designed for sequential data
- Maintains temporal dependencies (memory)
- Applications: time-series forecasting, NLP

### 4. Autoencoders
- Unsupervised learning
- Learns compressed representations
- Applications: anomaly detection, denoising

### 5. Generative Adversarial Networks (GAN)
- Two-network system (Generator vs Discriminator)
- Generates realistic synthetic data
- Applications: image generation, super-resolution

## Summary

| Architecture | Data Type | Core Strength | Use Case |
|-------------|----------|--------------|----------|
| ANN (MLP)   | Tabular  | Non-linear mapping | Regression, Classification |
| CNN         | Images   | Spatial feature extraction | Computer Vision |
| RNN/LSTM    | Sequence | Temporal memory | Time-series, NLP |
| Autoencoder | Unlabeled| Compression | Anomaly Detection |
| GAN         | Synthetic| Data generation | Image synthesis |

## Key Milestones

- **1950s**: Perceptron introduced
- **1969**: XOR problem → AI Winter
- **1986**: Backpropagation revived neural networks
- **1989**: CNN applied to digit recognition
- **2006**: Deep Learning re-emerged (Hinton)
- **2012**: AlexNet → breakthrough in image recognition

## Applications

- **Energy Systems**: Load forecasting, fault detection
- **Computer Vision**: Medical imaging, autonomous driving
- **NLP**: Translation, chat systems
- **Anomaly Detection**: Smart grid monitoring

## Conclusion

Neural networks are not a single solution but a collection of architectures.  
The choice of model depends on the type of data:

- Tabular → ANN  
- Images → CNN  
- Sequential → RNN/LSTM  

Understanding this mapping is essential before building any deep learning model.
